## Test Cartopy calls ##
Tests a couple different features of cartopy. When this cell is run, you should see a world map with all land masses colored green and country borderlines in black. 

```python
# Create and define the size of a figure object 
plt.figure(figsize=(8, 8))

# Create an axis with an basic PlateCarree projection style
ax = plt.axes(projection=ccrs.PlateCarree())
# Add coastlines
ax.coastlines('110m')
# Add land masses
ax.add_feature(cfeature.LAND, edgecolor="black", facecolor="green")
# Add borders
ax.add_feature(cfeature.NaturalEarthFeature(category='cultural',
                                           name='admin_0_countries',
                                           scale='10m',
                                           facecolor='none',
                                           edgecolor='black',
                                           linewidth=0.2));
```
